```python
import pandas as pd

# Ruta al archivo en su Google Drive
ruta_archivo = '/content/drive/MyDrive/Colab_Data/applicant_list_7.txt'

# Nombres de las columnas
nombres_solumnas = ['nombre', 'apellido', 'EPH', 'ECH', 'EMA', 'ECS', 'ESP', 'OP1', 'OP2', 'OP3']

# Leer un archivo con datos separados por espacio
df = pd.read_csv(ruta_archivo, sep=' ', names=nombres_solumnas)

# presentar el contenido de primeras líneas de DataFrame
display(df.head())
```

```python
# función para calcular valor de puntaje para admisión,
# formula según el departamento
def calcular_puntaje_admision(linea, opcion):
  departmento = linea[opcion]

  if departmento == 'Physics':
    puntajes = [linea['EPH'], linea['EMA'], linea['ESP']]
  elif departmento == 'Mathematics':
    puntajes = [linea['EMA'], linea['ECS'], linea['ESP']]
  elif departmento == 'Engineering':
    puntajes = [linea['EMA'], linea['ECS'], linea['EPH'], linea['ESP']]
  elif departmento == 'Chemistry':
    puntajes = [linea['ECH'], linea['EPH'], linea['ESP']]
  elif departmento == 'Biotech':
    puntajes = [linea['EMA'], linea['ECS'], linea['ECH'], linea['ESP']]
  else:
    return None  # Para caso cuando departamento no esta en la lista

  return round(sum(puntajes) / len(puntajes), 2)
```

```python
df['nombre_completo'] = df['nombre'] + ' ' + df['apellido']

df['OP1_score'] = df.apply(lambda r: calcular_puntaje_admision(r, 'OP1'), axis=1)
df['OP2_score'] = df.apply(lambda r: calcular_puntaje_admision(r, 'OP2'), axis=1)
df['OP3_score'] = df.apply(lambda r: calcular_puntaje_admision(r, 'OP3'), axis=1)

display(df.head())
```

```python
while True:
  try:
    N = int(input("Introduce el número máximo de alumnos en un departamento: "))
    break
  except ValueError:
    print("Entrada inválida. Por favor, introduce un número entero.")

print(f"Límite de admisión por departamento (N): {N}")
```

```python
departmentos = ["Biotech", "Chemistry", "Engineering", "Mathematics", "Physics"]

while True:
  departamento_seleccionado = input("Introduce el departamento para la selección (Biotech, Chemistry, Engineering, Mathematics, Physics): ")
  if departamento_seleccionado in departmentos:
    break
  else:
    print("Departamento inválido. Por favor, elige uno de la lista.")

print(f"Departamento seleccionado: {departamento_seleccionado}")
```

```python
# Filtrar el DataFrame para el departamento seleccionado y la opción OP1
aspirantes_departamento = df[(df['OP1'] == departamento_seleccionado)]

# Ordenar por OP1_score en orden descendente
aspirantes_ordenados = aspirantes_departamento.sort_values(by=['OP1_score'], ascending=[False])

# Seleccionar los N aspirantes con los puntajes más altos
aspirantes_elegidos = aspirantes_ordenados.head(N)

# Crear la tabla con las columnas especificadas
tabla_aspirantes_elegidos = pd.DataFrame({
    'nombre_completo': aspirantes_elegidos['nombre_completo'],
    'departmento': aspirantes_elegidos['OP1'],
    'puntaje': aspirantes_elegidos['OP1_score'],
    'opcion': 'OP1'
})

# Presentar el contenido de la tabla
display(tabla_aspirantes_elegidos)

# Imprimir el número de aspirantes elegidos
print(f"\nNúmero de aspirantes elegidos para {departamento_seleccionado} (OP1): {len(tabla_aspirantes_elegidos)}")
```
